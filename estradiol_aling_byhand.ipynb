```python
from ase.io import read, write
from ase.visualize import view
import numpy as np

atoms = read("geometry_estradiol.in", format="aims")
```

```python
#setting the COM

atoms.set_center_of_mass([0.0, 0.0, 0.0], scaled=False)
```

```python
import numpy as np

methyl = np.array([0.49803739, 2.62507462, 1.73111839])   # atom 17
carbon = np.array([0.83718779, 2.51748744, 0.23261283])   # atom 12

# Step 1: define the bond vector
vec = methyl - carbon  # vector from carbon to methyl

# Step 2: define target direction (z-axis)
target = np.array([0, 0, 1])

# Step 3: calculate angle using arccos (in radians)
def angle_between_vectors(v1, v2):
    v1_u = v1 / np.linalg.norm(v1)
    v2_u = v2 / np.linalg.norm(v2)
    dot = np.clip(np.dot(v1_u, v2_u), -1.0, 1.0)
    return np.arccos(dot)

angle_rad = angle_between_vectors(vec, target)
angle_deg = np.degrees(angle_rad)

print("Angle between methyl bond and z-axis:", angle_deg, "degrees")
```

```python
rot_axis = np.cross(vec, [0,0,1])

atoms.rotate(13.356998956280627, rot_axis, center='COM')
view(atoms)
print(atoms[17].position)
print(atoms[12].position)
```

```python
print(atoms[18].position)
print(atoms[5].position)
```

```python
# Positions of OH atoms
oxygen = np.array(atoms[18].position)   # atom 18
carbon = np.array(atoms[5].position)   # atom 5

# Bond vector from carbon to oxygen
vec2 = oxygen - carbon

# Project onto XY plane
vec_xy = np.array([vec2[0], vec2[1]])

# Calculate angle (in radians)
angle_rad = np.arctan2(vec_xy[1], vec_xy[0])
angle_deg = np.degrees(angle_rad)

print("Angle between OH vector and x-axis (XY-plane):", angle_deg, "degrees")
```
